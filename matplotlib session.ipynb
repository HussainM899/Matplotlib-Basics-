# **Matplotlib Library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
```

Basic Graph

```python
x = np.arange(-10,11)
y = x**2

plt.figure(figsize=(5,5))
plt.plot(x,y, "k*-")
plt.plot(x,-y, "*")
```

### Subplots

```python
x = np.arange(-10,11)
y = x**2

plt.figure(figsize=(10,10))

plt.subplot(2,2, 1)
plt.title("Plot 1")
plt.plot(x,y, "r", label = "curve")
plt.plot([0,0,0],[-10,0,100], label = "line")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")


plt.subplot(2,2, 2)
plt.title("Plot 2")
plt.plot(x,-y, "r", label = "curve")
plt.plot([0,0,0],[-100,0,10], label = "line")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(2,2, 3)
plt.title("Plot 3")
plt.plot([1,2,3,4],[1,2,3,4], label = "line")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(2,2, 4)
plt.title("Plot 4")
plt.plot(x,y, "--")
plt.plot([0,0,0],[-10,0,100])
```

```python
fig,ax = plt.subplots()
ax.plot(x,y)
ax.plot([0,0,0,0], [-10,20,30,100])
```

```python
fig, axes = plt.subplots(figsize = (5,5))

axes.plot(
    x,y, "red", linewidth = 3,
    marker = "o", markersize = 8, label = "X^2" 
    )

axes.plot(x, -1*y, "b--", label = "-X^2")

axes.set_xlabel ("X")
axes.set_ylabel ("X Squared")

axes.set_title ("My third plot")
axes.legend(loc = "lower center")
```

### Graph Plotting using NumPy Random Library 

```python
plt.figure(figsize=(10,10))
fig,( (ax1,ax2) , (ax3,ax4) ) = plt.subplots(2,2, figsize= (10,10))


ax1.set_title("Subplot 1")
ax2.set_title("Subplot 2")
ax3.set_title("Subplot 3")
ax4.set_title("Subplot 4")

ax4.plot(np.random.randn(50), c="yellow")
ax1.plot(np.random.randn(10), c="r")
ax3.plot(np.random.randn(15), c="blue")
ax2.plot(np.random.randn(100), c="black")

ax1.scatter (x, y, s = area, c = colors, alpha=1, cmap= "Spectral")
ax2.scatter (x, y, s = area, c = colors, alpha=1, cmap= "inferno")
```

### Histograms

```python
plt.subplots(figsize= (8,8))

plt.hist(values, bins=100, alpha = 0.5, histtype="bar", color="green", edgecolor = "steelblue")

plt.xlim(xmin = -5, xmax = +5)
```

### BoxPlot

```python
values = np.concatenate([np.random.randn(10), np.array([10,23.5,-15,-10])])

plt.figure(figsize = (8,4))
plt.boxplot (values)
```

### Pie Chart

```python
x= ['a','b','c','d']
y = [1,2,3,4]

color = ['r','g', 'b', 'y']

plt.figure()
plt.pie(y, labels = x, colors= color )
```

### SubPlot2Grid

```python
plt.figure(figsize=(10,6))

ax1 = plt.subplot2grid((3,3), (0,0), colspan=4)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=3)
ax4 = plt.subplot2grid((3,3), (2,0))
ax5 = plt.subplot2grid((3,3), (2,2))
```

```python
plt.figure()
plt.subplot2grid((3,2), (0,0), rowspan=2)

plt.subplot2grid((3,2), (0,1))
plt.subplot2grid((3,2), (1,1))
```
